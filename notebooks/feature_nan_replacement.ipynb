```python
import numpy as np
import pandas as pd

pd.options.display.max_columns = None
pd.set_option("display.max_rows", 100)
```

```python
def nan_integrity(df):
    df = df.replace("-", np.nan)
    print(df.columns)
    # at = df.loc[df.Name == "A"]
    at = df.loc[
        df.nucleotide == "A",
        (
            df.columns.str.startswith("bon_")
            | df.columns.str.startswith("ang_")
            | df.columns.str.startswith("atr_")
            | df.columns.str.startswith("aag_")
        ),
    ]
    atd = dict()
    for i in at.columns:
        atd[i] = at[i].isna().sum() / len(at)
    for key in list(atd.keys()):
        if atd[key] == 1.0:
            del atd[key]

    gt = df.loc[
        df.nucleotide == "G",
        (
            df.columns.str.startswith("bon_")
            | df.columns.str.startswith("ang_")
            | df.columns.str.startswith("atr_")
            | df.columns.str.startswith("aag_")
        ),
    ]

    gtd = dict()
    for i in gt.columns:
        gtd[i] = gt[i].isna().sum() / len(gt)
    for key in list(gtd.keys()):
        if gtd[key] == 1.0:
            del gtd[key]

    ct = df.loc[
        df.nucleotide == "C",
        (
            df.columns.str.startswith("bon_")
            | df.columns.str.startswith("ang_")
            | df.columns.str.startswith("atr_")
            | df.columns.str.startswith("aag_")
        ),
    ]
    ctd = dict()
    for i in ct.columns:
        ctd[i] = ct[i].isna().sum() / len(ct)
    for key in list(ctd.keys()):
        if ctd[key] == 1.0:
            del ctd[key]

    ut = df.loc[
        df.nucleotide == "U",
        (
            df.columns.str.startswith("bon_")
            | df.columns.str.startswith("ang_")
            | df.columns.str.startswith("atr_")
            | df.columns.str.startswith("aag_")
        ),
    ]
    utd = dict()
    for i in ut.columns:
        utd[i] = ut[i].isna().sum() / len(ut)
    for key in list(utd.keys()):
        if utd[key] == 1.0:
            del utd[key]
    common = set(list(atd.keys() & gtd.keys() & ctd.keys() & utd.keys()))
    return {
        "A": list(set(atd.keys()) - common),
        "C": list(set(ctd.keys()) - common),
        "G": list(set(gtd.keys()) - common),
        "U": list(set(utd.keys()) - common),
        "comm": list(common),
    }
```

```python
f = nan_integrity(pd.read_csv("/data/1/projects/master_thesis/rnaquanet/data/04_feature/ares_tiny/train_full_dataset.csv"))
```
