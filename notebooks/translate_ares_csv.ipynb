## Translate ares csv to RNAQuaNetStandard format

```python
import pandas as pd
import os

dir_path = '/home/adamczykb/Downloads/ares_1/'

target = pd.read_csv(f'{dir_path}target.csv',sep=' ')
train = [file for file in os.listdir(f'{dir_path}train') if os.path.isfile(os.path.join(f'{dir_path}train', file))]
target.loc[target['description'].isin(train)].to_csv(f'{dir_path}train.csv',sep=',',index=False)

test = [file for file in os.listdir(f'{dir_path}test') if os.path.isfile(os.path.join(f'{dir_path}test', file))]
target.loc[target['description'].isin(test)].to_csv(f'{dir_path}test.csv',sep=',',index=False)

val = [file for file in os.listdir(f'{dir_path}val') if os.path.isfile(os.path.join(f'{dir_path}val', file))]
target.loc[target['description'].isin(val)].to_csv(f'{dir_path}val.csv',sep=',',index=False)
```
