```python
import pandas as pd
from scipy import stats
import altair as alt
read_from_file = True
```

## Loading ares predictions

```python
if not read_from_file:
    original_trained_ares = pd.read_csv(
        "/home/adamczykb/rnaquanet/results/ares/output_ares.csv"
    )
    lociparse_trained_ares = pd.read_csv(
        "/home/adamczykb/rnaquanet/results/ares/output_trained_ares.csv"
    )
    original_trained_ares = original_trained_ares.rename(
        columns={"pred": "original_ARES_pred"}
    )
    lociparse_trained_ares = lociparse_trained_ares.rename(
        columns={"pred": "lociparse_ARES_pred"}
    )
    ares_comparison = pd.merge(
        original_trained_ares[["original_ARES_pred", "id"]],
        lociparse_trained_ares[["lociparse_ARES_pred", "id"]],
        on="id",
        how="outer",
    )[["id", "original_ARES_pred", "lociparse_ARES_pred"]]

    ares_comparison = ares_comparison.loc[
        ares_comparison["id"].isin(pd.unique(ares_comparison["id"]))
    ].set_index("id")
```

## Loading lociparse predictions

```python
if not read_from_file:
    original_trained_lociparse = pd.read_csv(
        "/home/adamczykb/rnaquanet/results/lociparse/lociparse_pretrained_rnapuzzle_results.csv"
    )
    lociparse_trained_lociparse = pd.read_csv(
        "/home/adamczykb/rnaquanet/results/lociparse/lociparse_selftrained_rnapuzzle_results.csv"
    )
    original_trained_lociparse = original_trained_lociparse.rename(
        columns={"error": "original_LOCIPARSE_pred_lDDT", "Name": "id"}
    )
    original_trained_lociparse["id"] = original_trained_lociparse["id"] + ".pdb"
    lociparse_trained_lociparse = lociparse_trained_lociparse.rename(
        columns={"error": "lociparse_LOCIPARSE_pred_lDDT", "Name": "id"}
    )
    lociparse_trained_lociparse["id"] = lociparse_trained_lociparse["id"] + ".pdb"
    lociparse_comparison = pd.merge(
        original_trained_lociparse[["original_LOCIPARSE_pred_lDDT", "id"]],
        lociparse_trained_lociparse[["lociparse_LOCIPARSE_pred_lDDT", "id"]],
        on="id",
        how="outer",
    )[["id", "original_LOCIPARSE_pred_lDDT", "lociparse_LOCIPARSE_pred_lDDT"]]

    lociparse_comparison = lociparse_comparison.loc[
        lociparse_comparison["id"].isin(pd.unique(lociparse_comparison["id"]))
    ].set_index("id")
```

## Loading RNAQuANet predictions

```python
if not read_from_file:
    lociparse_trained_rnaquanet_less_features = pd.read_csv(
        "/home/adamczykb/rnaquanet/results/rnapuzzle/rnaquanet_regular_loss_less_features/RNAQuANet_rnapuzzle_results.csv"
    )
    lociparse_trained_rnaquanet_contrastive_learned_less_features = pd.read_csv(
        "/home/adamczykb/rnaquanet/results/rnapuzzle/rnaquanet_modified_loss_less_features_contrastive/loss_rnapuzzle_results.csv"
    )
    lociparse_trained_rnaquanet_contrastive = pd.read_csv(
        "/home/adamczykb/rnaquanet/results/rnapuzzle/rnaquanet_modified_loss/LOCIPARSE_rnapuzzle_results.csv"
    )
    original_trained_rnaquanet_less_features = pd.read_csv(
        "/home/adamczykb/rnaquanet/results/rnapuzzle/rnaquanet_modified_loss/RNAQuANet_rnapuzzle_results.csv"
    )
    lociparse_trained_rnaquanet = pd.read_csv(
        "/home/adamczykb/rnaquanet/results/rnapuzzle/rnaquanet/LOCIPARSE_rnapuzzle_results.csv"
    )

    lociparse_trained_rnaquanet_less_features = (
        lociparse_trained_rnaquanet_less_features.rename(
            columns={
                "error": "lociparse_trained_RNAQuANet_less_features_pred",
                "Name": "id",
            }
        )
    )
    lociparse_trained_rnaquanet_less_features["id"] = (
        lociparse_trained_rnaquanet_less_features["id"] + ".pdb"
    )

    lociparse_trained_rnaquanet_contrastive_learned_less_features = lociparse_trained_rnaquanet_contrastive_learned_less_features.rename(
        columns={
            "error": "lociparse_trained_RNAQuANet_contrastive_learned_less_features_pred",
            "Name": "id",
        }
    )
    lociparse_trained_rnaquanet_contrastive_learned_less_features["id"] = (
        lociparse_trained_rnaquanet_contrastive_learned_less_features["id"] + ".pdb"
    )

    lociparse_trained_rnaquanet_contrastive = (
        lociparse_trained_rnaquanet_contrastive.rename(
            columns={"error": "lociparse_trained_RNAQuANet_contrastive_pred", "Name": "id"}
        )
    )
    lociparse_trained_rnaquanet_contrastive["id"] = (
        lociparse_trained_rnaquanet_contrastive["id"] + ".pdb"
    )

    original_trained_rnaquanet_less_features = (
        original_trained_rnaquanet_less_features.rename(
            columns={"error": "original_trained_RNAQuANet_less_features_pred", "Name": "id"}
        )
    )
    original_trained_rnaquanet_less_features["id"] = (
        original_trained_rnaquanet_less_features["id"] + ".pdb"
    )

    lociparse_trained_rnaquanet = lociparse_trained_rnaquanet.rename(
        columns={"error": "lociparse_trained_RNAQuANet_pred", "Name": "id"}
    )
    lociparse_trained_rnaquanet["id"] = lociparse_trained_rnaquanet["id"] + ".pdb"


    rnaquanet_comparison = pd.merge(
        lociparse_trained_rnaquanet[["lociparse_trained_RNAQuANet_pred", "id"]],
        original_trained_rnaquanet_less_features[
            ["original_trained_RNAQuANet_less_features_pred", "id"]
        ],
        on="id",
        how="outer",
    )
    rnaquanet_comparison = pd.merge(
        rnaquanet_comparison,
        lociparse_trained_rnaquanet_contrastive[
            ["lociparse_trained_RNAQuANet_contrastive_pred", "id"]
        ],
        on="id",
        how="outer",
    )
    rnaquanet_comparison = pd.merge(
        rnaquanet_comparison,
        lociparse_trained_rnaquanet_contrastive_learned_less_features[
            ["lociparse_trained_RNAQuANet_contrastive_learned_less_features_pred", "id"]
        ],
        on="id",
        how="outer",
    )
    rnaquanet_comparison = pd.merge(
        rnaquanet_comparison,
        lociparse_trained_rnaquanet_less_features[
            ["lociparse_trained_RNAQuANet_less_features_pred", "RMSD", "length", "id","Lab","PZ"]
        ],
        on="id",
        how="outer",
    )

    rnaquanet_comparison = rnaquanet_comparison.loc[
        rnaquanet_comparison["id"].isin(pd.unique(rnaquanet_comparison["id"]))
    ].set_index("id")
```

```python
if not read_from_file:
    comparesion = pd.merge(
        ares_comparison,
        lociparse_comparison,
        left_index=True,
        right_index=True,
        how="outer",
    )
    comparesion = pd.merge(
        comparesion,
        rnaquanet_comparison,
        left_index=True,
        right_index=True,
        how="outer",
    )
    comparesion.to_csv("/home/adamczykb/rnaquanet/results/comparesion.csv")
    comparesion['original_LOCIPARSE_pred_lDDT_rev']=1.0-comparesion['original_LOCIPARSE_pred_lDDT']
    comparesion['lociparse_LOCIPARSE_pred_lDDT_rev']=1.0-comparesion['lociparse_LOCIPARSE_pred_lDDT']
```

## Spearman correlation for each puzzle

```python
translate={
    "original_ARES_pred": "ARES",
    "lociparse_ARES_pred": "LOCIPARSE ARES",
    "original_LOCIPARSE_pred_lDDT_rev": "LOCIPARSE",
    "lociparse_LOCIPARSE_pred_lDDT_rev": "LOCIPARSE self-trained",
    "lociparse_trained_RNAQuANet_pred": "RNAQuANet, trained using LOCIPARSE",
    "lociparse_trained_RNAQuANet_less_features_pred": "RNAQuANet limited features, trained using LOCIPARSE",
    "original_trained_RNAQuANet_less_features_pred": "RNAQuANet limited features, trained using RNAQuANet",
    "lociparse_trained_RNAQuANet_contrastive_pred": "RNAQuANet contrastive train, trained using LOCIPARSE",
    "lociparse_trained_RNAQuANet_contrastive_learned_less_features_pred": "RNAQuANet contrastive train with limited features,trained using LOCIPARSE",
}
```

```python
pd.set_option("display.max_colwidth", None)
import altair as alt
model_ranking = {model: [] for model in translate.values()}
for PZ in pd.unique(comparesion["PZ"]):
    display(PZ)
    result = {"model": [], "spearman": [], "MAE": [], "MSE": []}
    result_pd = []
    for col in translate.keys():
        data_pz = comparesion[comparesion["PZ"] == PZ].copy()
        data_pz = data_pz.sort_values(by="RMSD").reset_index( drop=True)
        rmsd_sorted = data_pz.copy().sort_values(by="RMSD").index
        error_sorted = data_pz.copy().sort_values(by=col).index
        result_pd.append(
            pd.DataFrame(
                {
                    "RMSD rank": rmsd_sorted,
                    "prediction rank": error_sorted,
                    "series": translate[col],
                }
            )
        )
        result["model"].append(translate[col])
        result["spearman"].append(stats.spearmanr(rmsd_sorted, error_sorted).statistic)
        result["MAE"].append((data_pz["RMSD"] - data_pz[col]).abs().mean())
        result["MSE"].append(((data_pz["RMSD"] - data_pz[col]) ** 2).mean())
    highlight = alt.selection_point(fields=["series"], bind="legend")
    chart = (
        alt.Chart(pd.concat(result_pd, ignore_index=True),width=1000, height=1000)
        .mark_circle(size=80)
        .encode(
            x="RMSD rank",
            y="prediction rank",
            color="series:N",
            opacity=alt.condition(highlight, alt.value(1), alt.value(0.1)),
            tooltip=["RMSD rank", "prediction rank", "series"],
        )
        .add_params(highlight)
        .properties(
            title="Ranking [PUZZLE {:s}] ".format(
                PZ,
            )
        )
    ).interactive()
    print(pd.DataFrame(result).sort_values("spearman", ascending=False).to_string())
    for index,values in pd.DataFrame(result).sort_values("spearman", ascending=False).reset_index().iterrows():
        model_ranking[values['model']].append(index+1)
    chart.show()
    display("==========================================")
```

```python
data = pd.DataFrame([
    {'group': group, 'position': pos}
    for group, positions in model_ranking.items()
    for pos in positions
])
chart = alt.Chart(data, width=700,height=700).transform_bin(
    'binned_position', field='position', bin=alt.Bin(step=1, extent=[1, 10])
).transform_aggregate(
    count='count()',
    groupby=['group', 'binned_position']
).transform_impute(
    impute='count',
    key='binned_position',
    groupby=['group'],
    value=0
).mark_bar(opacity=1,size=75).encode(
    x=alt.X('binned_position:N', title='Contest Position'),
    y=alt.Y('count:Q', title='Count'),
    color='group:N'
).properties(
    title='Model Performance Histogram (Lower = Better)',
)

chart.show()
```
